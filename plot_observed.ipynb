```python
import pandas as pd
```

```python
data = pd.read_excel("data/monitoring.xlsx")
```

```python
data.columns
```

```python
counts = data['well_num_id'].value_counts() 
```

```python
counts[counts > 30].index
```

```python

data_filtered = data[data['well_num_id'].isin(counts[counts > 30].index)]
```

```python
# lets take the first well 

first_well = data_filtered['well_num_id'].iloc[10]

data_per_well = data_filtered[data_filtered['well_num_id'] == first_well]
```

```python
data_per_well
```
